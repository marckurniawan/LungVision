```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../dataset/lung_cancer_dataset.csv")

df.head()
```

```python
print(df.shape)

print(df.info())

df.describe()
```

```python
missing = df.isnull().sum()

missing

plt.figure(figsize=(12,5))
missing.plot(kind="bar")

plt.title("Missing Value per Feature")
plt.xlabel("Feature")
plt.ylabel("Jumlah Missing Value")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
df["Result"].value_counts()

plt.figure(figsize=(6,4))

sns.countplot(data=df, x="Result")

plt.title("Target Distribution")

plt.show()
```

```python
numeric_columns = df.select_dtypes(include=np.number).columns

df[numeric_columns].hist(
    figsize=(18,15),
    bins=20
)

plt.show()
```

```python
plt.figure(figsize=(18,12))

df.boxplot(rot=90)

plt.show()
```

```python
plt.figure(figsize=(18,12))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap="coolwarm",
    annot=False
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
corr = df.corr(numeric_only=True)["Result"].sort_values(ascending=False)

corr

plt.figure(figsize=(10,6))

corr.drop("Result").plot(kind="bar")

plt.title("Correlation with Target")

plt.show()
```

```python
selected = [
    "Age",
    "Smoking",
    "Genetic Risk",
    "Air Pollution",
    "Result"
]

sns.pairplot(
    df[selected],
    hue="Result"
)

plt.show()
```

```python
sns.countplot(
    data=df,
    x="Gender",
    hue="Result"
)

plt.show()
```

```python
# Jumlah setiap kelas
print(df["Result"].value_counts())
print(df["Result"].value_counts(normalize=True) * 100)

plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x="Result"
)

plt.title("Distribusi Target")
plt.xlabel("Result")
plt.ylabel("Jumlah Data")

plt.show()

df["Result"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.ylabel("")
plt.title("Persentase Target")

plt.show()
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

df[numeric_cols].hist(
    figsize=(18, 15),
    bins=20,
    edgecolor="black"
)

plt.suptitle("Distribusi Setiap Fitur", fontsize=18)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10))

df.boxplot(rot=90)

plt.title("Boxplot Semua Fitur")

plt.tight_layout()

plt.show()
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()} nilai unik")
```

```python
important_features = [
    "Smoking",
    "Age",
    "Air Pollution",
    "Genetic Risk"
]

for feature in important_features:

    plt.figure(figsize=(6,4))

    sns.histplot(df[feature], bins=10)

    plt.title(feature)

    plt.show()
```

```python
# Hitung korelasi antar fitur numerik
corr_matrix = df.corr(numeric_only=True)

# Visualisasi heatmap
plt.figure(figsize=(18, 12))

sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    annot=False,
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
corr_target = (
    df.corr(numeric_only=True)["Result"]
      .sort_values(ascending=False)
)

print(corr_target)
```

```python
plt.figure(figsize=(12,6))

corr_target.drop("Result").plot(kind="bar")

plt.title("Correlation of Features with Target")
plt.xlabel("Feature")
plt.ylabel("Correlation")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
